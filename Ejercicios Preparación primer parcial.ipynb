# Ejericios



## Ciclos


*   Escriba un programa que dado un string de numeros separaados por espacios, imprima todos los numeros positivos.
*   Escriba un programa que diga si un numero es primo no.
*  Escriba un programa para saber si un string es un palindromo.
*  Escriba un programa que imprima todos los numeros primos que se encuentran dentro de los primeros n numeros.
* Escriba un programa que reciba un conjunto de numeros separados por espacio, un numero a buscar y encuentre todas las ocurrencias.
* Dado un string de numeros separados por espacio encuentre el numero minimo y maximo.
* Escriba un programa para calcular la siguiente sumatoria: a + a$r$ + a$r^2$ + a$r^3$ + ……… + a$r^n = \sum_{i=0}^{n} ar^i$ 
* Escribir un programa que reciba una dos numeros y calcule la suma de los numeros pares entre esos numeros.




## Cadena de caracteres

*   Escriba un programa que dada una cadena de caracteres, retorne la cantidad de caracteres.
*   Escriba un program que dado un string retorne un nuevo string con los dos primeros caracteres concatenado con los dos ultimos, si el string es menos de longitud 4, devolver vacio.
* Escriba un programa que dado un string y un caracter remplace todas las ocrrecias de este por el caracter '\$'.
* Escriba un programa que dado $i^{th}$ caracter lo remueva de un string.
* Escriba un programa que dado que remueva todos los caracteres impares de un string.
* Escriba un programa que retorne un string hasta antes de un caracter especificado.
* Escriba un programa que cuente la cantidad de vocales de un string.
* Escriba un programa que cuente la maxima cantidad de ceros consecutivos en un string de un numero binario.
* Escriba un programa para remover todos los caracteres de un string, excepto por el especficado.
* Escriba un programa para encontar la palabra mas corta y la mas larga de un string.
