# Identificadores
Un identificador es una secuencia de símbolos que se utilizan como nombres de variables, funciones, arreglos, clases y otras estructuras de los lenguajes de programación.



```
my_number = 4  my_number is of type int
my_string = "Sally" my_string is of type str
my_decimal = 3.0 my_decimal is of type float
my_bool = True my_bool is of type bool
```



````python
my_number = 4
print(my_number, type(my_number), sep=" ")
my_string = "Andres"
print(my_string, type(my_string), sep=" ")
my_decimal = 3.0
print(my_decimal, type(my_decimal), sep=" ")
my_bool = True
print(my_bool, type(my_bool), sep=" ")
````

# Operadores

## Aritmeticos
Para los datos de tipo numérico se pueden utilizar los siguientes:


1.   Suma: +
2.   Resta: -
3.   Multiplicación: *
4.   División: /
5.   División piso: //
6.   Resto: %
7.   Potencia: **




````python
suma = x + y
print("La suma es %d" % (suma))
resta = x - y
print("La resta es %d" % (suma))
x_negative = -y
print("El valor es %d" % (suma))
multipl = x * y
print("La multiplicacion es %d" % (multipl))
div = x / k
print("La división es %.2f" % (div))
div_entera = x // k
print("La división es", div_entera)
modulo = x % k
print("El resto de la división es %d" % (modulo))
potencia = k ** 3
print("El numero %d elevado a la potencia 3 es %d" % (k, potencia))
````

## Asignación
Para asignar valores a variables se pueden utilizar los siguientes operadores:



1.   Asignación: =
2.   Asignación con suma: +=
3.   Asiganción con resta: -=
4.   Asignación con multiplicación: *=
5.   Asignación con división: /=
6.   Asignación con división entera: //=
7.   Asignación con división residuo: %=
8.   Asignación con potencia: **=





````python
m = 10
n = 3
m += 1 # es igual a m = m + 1
print(m)
m -= 2 # es igual a m = m - 2
print(m)
m *= 2 # es igual a m = m * 2
print(m)
m /= n # es igual a m = m / n
print(m)
m = 10
m //= n # es igual a m = m // n
print(m)
m **= n # es igual a m = m ** n
print(m)
m %= n # es igual a m = m % n
print(m)
````

## Logicos



1.   not: Operador ¬ de la negación en Python. Si la expresión α es falsa su negación es verdadera y si es verdadera su negación es falsa.
2.   and: Operador ∧ de la conjunción en Python. La expresión
α ∧ β solo es verdadera cuando ambas α y β son verdaderas. En otro caso es falsa.
3.   or: Operador ∨ de la disyunción en Python. La expresión
α ∨ β solo es falsa cuando ambas α y β son falsas. En otro caso es verdadera.



````python
v1 = True
v2 = False

## For not
print(not v1)
print(not v2)
## For And
print(v1 and v1)
print(v1 and v2)
print(v2 and v1)
print(v2 and v2)
## For Or
print(v1 or v1)
print(v1 or v2)
print(v2 or v1)
print(v2 or v2)
````

## Igualdad y relaciones


1.   $==$ : Devuelve Verdadero si dos valores son iguales.
2.   $!=$ : Devuelve Verdadero si dos valores son distintos.
3.   $>$ : Mayor que, devuelve Verdadero si el primer
operando es estrictamente mayor que el segundo.
4.   $<$ : Menor que, devuelve Verdadero si el primer operando es estrictamente menor que el segundo.
5.   $>=$ : Mayor o igual, devuelve Verdadero si el primer operando es mayor o igual que el segundo.
6.   $<=$ : Menor igual, devuelve Verdadero si el primer operando es menor o
igual que el segundo.





````python
x1 = 10
x2 = 4
x3 = 10

print(x1 == x3)
print(x1 != x2)
print(x1 > x2)
print(x1 < x2)
print(x1 >= x3)
print(x1 <= x3)
````

# Evaluación de expresiones

Como se vio previamente las cadenas =, +=, -=, *=, /=, //=, %= sirven para representar los operadores de asignación, es decir, permiten asignar un valor a una determinada variable, donde la variable se encuentra a la izquierda del operador y el valor resulta de evaluar una expresión que se encuentra a la derecha del operador.

Se debe tener cuidado que al evaluar una expresión el resultado sea del mismo tipo de la variable a la que se le esta asignando el valor, esto es, que a una variable de tipo int se le asigne un valor entero, que a una variable de tipo float se le asigne un valor real, que a una variable de tipo bool se le asigne un valor booleano, que a una variable de tipo str se le asigne una cadena, etc. Python no controla esto.
